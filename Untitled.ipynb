```python
import sys
import os
import platform
import psutil
import pkg_resources  # To list installed libraries

# Display Python version
print(f"Python version: {sys.version}")

# Display the location of the Python executable
print(f"Python executable location: {os.path.abspath(sys.executable)}")

# Display the operating system and architecture
print(f"Operating System: {platform.system()} {platform.release()} ({platform.architecture()[0]})")

# Display the installed libraries and their versions using pkg_resources
print("\nInstalled libraries:")
for dist in pkg_resources.working_set:
    print(f"{dist.project_name}=={dist.version}")

# Display CPU information
cpu_count = psutil.cpu_count(logical=False)  # Physical cores
print(f"\nCPU: {cpu_count} physical cores")

# Display Memory information
memory = psutil.virtual_memory()
print(f"Total RAM: {memory.total / (1024 ** 3):.2f} GB")

```

```python
import mlflow
import mlflow.sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Define experiment name
experiment_name = "Iris_Classification_Experiment"

# Set the experiment name
mlflow.set_experiment(experiment_name)

# Define run name
run_name = "RandomForest_Run_1"

# Load dataset
data = load_iris()
X = data.data
y = data.target

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Start the MLflow run with a specific run name
with mlflow.start_run(run_name=run_name):

   # Model training
   model = RandomForestClassifier()
   model.fit(X_train, y_train)

   # Create input example (sample input data)
   input_example = X_train[0:1]

   # Log the model with input example
   mlflow.sklearn.log_model(model, "random_forest_model", input_example=input_example)

   # Log parameters (optional)
   mlflow.log_param("n_estimators", model.n_estimators)

   # Log metrics (optional)
   mlflow.log_metric("accuracy", model.score(X_test, y_test))

   # Print out the current logged details of the MLflow run
   run_id = mlflow.active_run().info.run_id
   print(f"Run ID: {run_id}")

   # Fetch the logged parameters
   params = mlflow.get_run(run_id).data.params
   print("Logged Parameters:")
   for param in params:
       print(f"{param}: {params[param]}")

   # Fetch the logged metrics
   metrics = mlflow.get_run(run_id).data.metrics
   print("Logged Metrics:")
   for metric in metrics:
       print(f"{metric}: {metrics[metric]}")

   # Fetch the logged model artifact URI
   artifact_uri = mlflow.get_artifact_uri()
   print(f"Model Artifact URI: {artifact_uri}")

# End the run
mlflow.end_run()

```
