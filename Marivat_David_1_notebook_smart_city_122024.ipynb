```python
# Importing the pandas and numpy library
import pandas as pd
import numpy as np

# This function reads a csv file and imports it into a variable
dataTree = pd.read_csv('data/arbres.csv')


# Calculating indicators
average = dataTree.mean(numeric_only=True)
medians = dataTree.median(numeric_only=True)

numericColumns = dataTree.select_dtypes(include=[np.number])
quantiles = numericColumns.quantile([0.25, 0.5, 0.75]) # 25%, 50%, et 75% of datas

# Display of results
print("="*20 + " Résultats " + "="*20)
print("Moyennes :\n", average)
print("\nMédianes :\n",medians)
print("\nQuantiles :\n",quantiles)
```

<h4><u>Discrete quantitative variables</u></h4>

<h6><u>Circonference_cm variables</u></h6>

```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt

# Lire les données
dataTree = pd.read_csv('data/arbres.csv')

# Box plot
dataTree.boxplot(column='circonference_cm')

# Calculer la moyenne, la médiane et l'écart-type
mean = dataTree['circonference_cm'].mean()
median = dataTree['circonference_cm'].median()
std = dataTree['circonference_cm'].std()

# Ajouter la ligne de la moyenne
plt.axhline(y=mean, color='r', linestyle='--', label='Moyenne')

# Ajouter la ligne de la médiane
plt.axhline(y=median, color='b', linestyle='-', label='Médiane')

# Ajouter les lignes de l'écart-type
plt.axhline(y=mean+std, color='g', linestyle=':', label='Écart-type')
plt.axhline(y=mean-std, color='g', linestyle=':')

# Ajouter des annotations textuelles avec des positions ajustées
# Moyenne (au-dessus)
plt.annotate(f'Moyenne: {mean:.2f}', 
             xy=(0.5, mean + 0.05), xycoords='data',  # Légèrement au-dessus de la ligne
             bbox=dict(boxstyle="round,pad=0.3", edgecolor='red', facecolor='white'),
             color='red', fontsize=10, ha='center')

# Médiane (au-dessous)
plt.annotate(f'Médiane: {median:.2f}', 
             xy=(0.5, median - 0.05), xycoords='data',  # Légèrement en dessous de la ligne
             bbox=dict(boxstyle="round,pad=0.3", edgecolor='blue', facecolor='white'),
             color='blue', fontsize=10, ha='center')

# Écart-type (à droite)
plt.annotate(f'Écart-type: {std:.2f}', 
             xy=(1.05, mean + std), xycoords='data',  # À droite de la ligne
             bbox=dict(boxstyle="round,pad=0.3", edgecolor='green', facecolor='white'),
             color='green', fontsize=10, ha='left')

# Titre et labels
plt.title('Box Plot circonference_cm')
plt.ylabel('Circonference_cm')

# Ajouter une légende
plt.legend()

# Ajuster la mise en page
plt.tight_layout()

# Afficher le graphique
plt.show()
```

<h1 style="text-align: center; font-weight: bold;"><u>Step 4 Detect and process outliers in the dataset</u></h1>

<h2 style="font-weight: bold;"><u>z-score technique</u></h2>

```python
import pandas as pd
import numpy as np

# This function reads a csv file and imports it into a variable
dataTree = pd.read_csv('data/arbres.csv')

# Calcul de la moyenne et de l'écart-type
mean = dataTree['circonference_cm'].mean()
std_dev = dataTree['circonference_cm'].std()

# Calcul des z-scores
dataTree['z_score'] = (dataTree['circonference_cm'] - mean) / std_dev

# Détection des outliers
threshold = 3  # Change à 2 si tu veux une détection plus souple
outliers = dataTree[abs(dataTree['z_score']) > threshold]

# Affichage des résultats
print("Moyenne :", mean)
print("Écart-type :", std_dev)
print("Outliers détectés :")
print(outliers['circonference_cm'])
```
